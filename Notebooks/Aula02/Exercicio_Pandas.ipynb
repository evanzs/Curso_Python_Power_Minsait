```python
import pandas as pd
```

```python
df_pandas_2json = pd.read_json("https://raw.githubusercontent.com/thaisratis/Curso_Python_Power_Minsait/main/Datasets/Exercicio_Pandas_2.json")
df_pandas_2json.head()


```

```python
df_pandas_3csv = pd.read_csv("https://raw.githubusercontent.com/thaisratis/Curso_Python_Power_Minsait/main/Datasets/Exercicio_Pandas_3.csv",sep="\t")
df_pandas_3csv.head()
```

```python

df_combined =pd.concat([df_pandas_3csv,df_pandas_2json],axis=0)

df_combined.head()

df_groupy = df_combined.groupby(['Produto','Valor_Unitario']).agg(['sum', 'mean'])
df_groupy.head()
```
